```python
# tipo, adresse, code postal

tipo=[]
adresse=[]
code=[]
for p in soup.find_all('span', attrs={"itemprop" : "address"}):    
    dois = p.text.split('au')
    tipo.append(dois[0].replace("\n","").strip())
    adresse.append(dois[1].replace("\n","").strip())
    code.append(dois[1].split(",")[1].strip().split(" ")[0])

print(tipo)
print(adresse)
print(code)
```

```python
# horaire de fermeture

time=[]
for elem in soup.findAll('div', attrs={"class" : "feature-content"})[3]:
    if '\n' in elem:
        next
    else:
        time.append(elem.text)

print(time)
```

```python
time=[]
for elem in soup.findAll('div', attrs={"class" : "feature-content"})[3]:
    print(elem)
    for p in elem('p'):
        time.append(p.text)

#print(time)
```
